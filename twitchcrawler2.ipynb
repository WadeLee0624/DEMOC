```python
from bs4 import BeautifulSoup
import requests
import re  
import csv
import math
import time
import threading
import datetime
from pandas import DataFrame
from pandas import read_csv
from selenium import webdriver  #載入driver
from selenium.common.exceptions import NoSuchElementException
from selenium.webdriver.common.action_chains import ActionChains #載入滑鼠
from selenium.webdriver.common.keys import Keys #載入滑鼠


headers = {
    'User-Agent':"Mozilla/5.0 (Windows NT 6.1; WOW64) AppleWebKit/537.1 (KHTML, like Gecko) Chrome/22.0.1207.1 Safari/537.1"
}  


def top(n):
    print("呼叫TOP函數")               #此函式找出目前觀看人數前幾名LOL實況，回傳開台者名單
    driver = webdriver.Chrome('D:/driver/chromedriver') #要先下載chrome driver，並給予儲存路徑
    driver.set_window_size(1000,5000)
    onstreaming = []                 #儲存正在直播的實況間代碼
    url="https://www.twitch.tv/directory/game/League%20of%20Legends?sort=VIEWER_COUNT"  
    driver.get(url)
    time.sleep(3)
    search_filter = driver.find_element_by_xpath('//input[@placeholder="搜尋標籤"]')   #搜尋英文標籤，目的是只抓英文頻道
    search_filter.send_keys(u"英文")
    time.sleep(1)
    Englinshlabel = driver.find_element_by_xpath('//button[@tabindex="-1"]')  #點選切換英文標籤
    Englinshlabel.click()
    time.sleep(1)
    Sortlabel = driver.find_element_by_xpath('//button[@data-a-target="browse-sort-menu"]')
    Sortlabel.click()  
    driver.execute_script('window.scrollTo(0, document.body.scrollHeight);')
    top_soup = BeautifulSoup(driver.page_source,'lxml')  #第一次寫入soup
    streamer = top_soup.find_all('a',attrs={"data-a-target": "preview-card-title-link"})
    for i in range(n):
        onstreaming.append(streamer[i].attrs['href'].split("/",2)[1])
        #print(streamer[i].attrs['href'].split("/",2)[1])
    driver.close()
    return onstreaming


def crawler(n):                              #此函數執行tread，啟動函式為crawler_tread，參數n為同時抓取的直播台數量
    print("呼叫crawler函數")
    num = n
    steamer_list = top(20)
    for i in steamer_list:
        if i not in in_thread.keys() and num>0 :  
            in_thread[i] = 1
            threading.Thread(target = crawler_tread, args = (i,)).start()        
            num -= 1
           
    
def crawler_tread(streamer):                #此函數執行tread，啟動函式為crawlering，確認直播台是否正在開台，若為是則拜訪，
    while (in_thread.get(streamer) == 1):
        threading.Thread(target = crawlering, args = (streamer,)).start()
        time.sleep(60)

        
def crawlering(streamer):                    # 對網頁爬取的函數
    view_inf = []
    view_inf.append(time.strftime("%Y-%m-%d %H:%M:%S", time.localtime()))
    url = "https://www.twitch.tv/"+streamer
    driver1 = webdriver.Chrome('D:/driver/chromedriver')
    driver1.get(url)
    time.sleep(3)
    soup = BeautifulSoup(driver1.page_source,'lxml')  #第一次寫入soup
    driver1.find_element_by_xpath('//button[@aria-label="關閉「訂閱九是爽」訊息"]').click() #點擊觀眾按鈕
    try:
        streamer_s = soup.find('div',attrs={"data-a-target":"channel-viewers-count"}) 
        viewer_cnt = soup.find('div',attrs={"data-a-target":"tw-stat-value"})
        viewer_cnt = viewer_cnt.string.replace(',','')     
        iftoomanyv = int(viewer_cnt)>18000
        if streamer_s == None or iftoomanyv :
            driver1.close()
            in_thread[streamer] = 0
            crawler(1)
        driver1.find_element_by_xpath('//button[@aria-label="聊天室中的使用者"]').click() #點擊觀眾按鈕
        time.sleep(3)
        roll_list = driver1.find_element_by_xpath("//div[@class='simplebar-scroll-content chat-viewers__scroll-container']") #找到觀眾list卷軸
        roll_list.click()  #先讓鼠標點擊一下
        x = math.log(int(viewer_cnt))*100     #特殊字元替換後用人數/25看要跑幾圈 
        for i in range(1,int(x)):  #此迴圈滾動觀眾在線清單畫面
            ActionChains(driver1).key_down(Keys.PAGE_DOWN).perform()  
            time.sleep(0.05)
        soup = BeautifulSoup(driver1.page_source,'lxml')    #在寫入一次soup覆蓋，因為滾動過
        driver1.close()
        soup1 = soup.find('div',attrs={"aria-labelledby": "chat-viewers-list-header-使用者"})  #soup1找到使用者清單位置
        viewer = soup1.find_all('button',class_='chat-viewers-list__button')   #所有按鈕中的使用者id寫入viewer
        today = str(datetime.date.today()).replace("-","")
        csv_name = streamer +"_"+today+".csv"
        view_inf.append(viewer_cnt)
        for x in viewer:   
            view_inf.append(x.string) 
        with open(csv_name,'a',newline = '') as f:
            csv_write = csv.writer(f)
            csv_write.writerow(view_inf)
        f.close()
    except:
        try:
            streamer_s = soup.find('p',class_= "tw-c-text-live tw-strong") 
            viewer_cnt = streamer_s.text.replace(',','')
            iftoomanyv = int(viewer_cnt)>18000
            if streamer_s == None or iftoomanyv :
                driver1.close()
                in_thread[streamer] = 0
                crawler(1)           
        except:
            driver1.close()
            in_thread[streamer] = 0
            crawler(1)
        else:
            view_inf.append(viewer_cnt)
            driver1.find_element_by_xpath('//button[@aria-label="聊天室中的使用者"]').click() #點擊觀眾按鈕
            time.sleep(3)
            roll_list = driver1.find_element_by_xpath("//div[@class='simplebar-scroll-content chat-viewers__scroll-container']") #找到觀眾list卷軸
            roll_list.click()  #先讓鼠標點擊一下
            x = math.log(int(viewer_cnt))*60   #特殊字元替換後用人數/25看要跑幾圈 
            for i in range(1,int(x)):  #此迴圈滾動觀眾在線清單畫面
                ActionChains(driver1).key_down(Keys.PAGE_DOWN).perform()  
                time.sleep(0.05)
            soup = BeautifulSoup(driver1.page_source,'lxml')    #在寫入一次soup覆蓋，因為滾動過
            driver1.close()
            soup1 = soup.find('div',attrs={"aria-labelledby": "chat-viewers-list-header-使用者"})  #soup1找到使用者清單位置
            viewer = soup1.find_all('button',class_='chat-viewers-list__button')   #所有按鈕中的使用者id寫入viewer
            today = str(datetime.date.today()).replace("-","")
            csv_name = streamer +"_"+today+".csv"  
            for x in viewer:   
                view_inf.append(x.string) 
            with open(csv_name,'a',newline = '') as f:
                csv_write = csv.writer(f)
                csv_write.writerow(view_inf)
            f.close()





#此為主程式

in_thread = dict()
threads = []  
crawler(3)





```
